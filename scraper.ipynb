```python
from bs4 import BeautifulSoup
import requests
import time
import re
import json
```

# Bash.org scrapper 
## Third version

```python
def scrape(file = "scrape.csv", page_number = 1):
    success = True
    file_name, file_extension  = file.split(sep=".")
    if file_extension == "json":
        json_dump = []
    with open(file, "w") as f:
        while success:
            r = requests.get("http://bash.org/?browse=" + str(page_number), stream=True)
            soup = BeautifulSoup(r.text, 'html.parser')
            # sanity check
            if not soup.get_text().find("Invalid page.") == -1:
                print("Invalid page")
                success = False
            elif soup.title.contents[0] == "404 Not Found":
                print("404 Not Found")
                success = False
            elif soup.find_all("p", "quote") == []:
                print("empty page")
            else:
                print("success, page", page_number)

            quote_infos = soup.find_all("p", "quote")
            quotes = soup.find_all("p", "qt")

            # more sanity checks
            assert len(quote_infos) == len(quotes)
            

            for quote_info, quote in zip(quote_infos, quotes):
                if file_extension == "json":
                    json_dump_quote = []
                quote_number = quote_info.b.get_text()
                score = quote_info.font.get_text()
                sentences = quote.get_text().splitlines()
                if file_extension == "json":
                    json_dump_quote.append(quote_number)
                    json_dump_quote.append(score)
                    json_dump_sentences = []
                for sentence in sentences:
                    try:
                        username = re.findall("<.*?>", sentence)[0]
                        sentence = re.split("<.*?> ", sentence, 1)[1]
                        sentence = sentence.replace("\"", "\"\"")
                    except IndexError:
                        username = "-"
                        sentence = sentence.replace("\"", "\"\"")
                    if file_extension == "csv":
                        f.write(f"\"{quote_number}\", \"{score}\", \"{username}\", \"{sentence}\"\n")
                    elif file_extension == "json":
                        json_dump_sentences.append([username, sentence])
                if file_extension == "json":
                    json_dump_quote.append(json_dump_sentences)
                    json_dump.append(json_dump_quote)
            page_number += 1
        json.dump(json_dump, f)
```

```python
scrape("third scrape.json")
```
